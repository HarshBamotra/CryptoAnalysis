## Method 1: Using miners revenue and hashrate % of countries

Importing required libraries

```python
import numpy as np
import pandas as pd
```

Reading the following csv files:
1. miners-revenue.csv ----- Bitcoin mining revenue data
2. export.csv ------------- Country wise hashrate %

```python
revenueDF = pd.read_csv("miners-revenue.csv")
miningDF = pd.read_csv("export.csv")

revenueDF.Timestamp = pd.to_datetime(revenueDF.Timestamp , format="%Y%m%")
miningDF.date = pd.to_datetime(miningDF.date , format="%Y%m%")
revenueDF
miningDF
```

Extracting all revenues of year 2020 and handling missing values

```python
df = revenueDF[revenueDF['Timestamp'].dt.year == 2020].copy()
df2 = pd.DataFrame()
df2["Timestamp"] = pd.date_range(start ='01-01-2020', end ='31-12-2020', freq ='24H')
for i in df.index:
    for j in df2.index:
        if(df.loc[i , "Timestamp"] == df2.loc[j , "Timestamp"]):
            df2.loc[j , "miners-revenue"] = df.loc[i , "miners-revenue"]
df = df2.fillna(method = "ffill")
df = df.fillna(method = "bfill")
df
```

Calculating monthly revenue 

```python
df = df.groupby(df["Timestamp"].dt.month  , as_index=False).agg({"miners-revenue" : "sum"})
df["Month"] = ['2020-01-01', '2020-02-01', '2020-03-01', '2020-04-01', '2020-05-01', '2020-06-01', '2020-07-01', '2020-08-01', '2020-09-01', '2020-10-01', '2020-11-01', '2020-12-01']
df.Month = pd.to_datetime(df.Month , format="%Y%m%")
df
```

Extracting hashrate data of year 2020

```python
df1 = miningDF[miningDF['date'].dt.year == 2020].copy()
df1 = df1.reset_index(drop=True)
df1["Spend on electricity"] = 0
df1["Average electicity price"] = 0
df1["monthly_hashrate_%"] = (df1["monthly_hashrate_%"].str.replace(r"\D" , "")).astype(int , copy=True, errors='raise')/100
df1
```

Calculating cost spend on electricity by assuming that an average of 60% of the total revenue is spend on electricity in USDT.

```python
df['Spend on electricity'] = df["miners-revenue"]*(0.6)
df
```

```python
for i in df.index:
    for j in df1.index:
        if(df.iloc[i , 1]==df1.iloc[j , 0]):
            df1.iloc[j , 4] = (df.iloc[i , 2]*(df1.iloc[j , 2]))/100
df1
```

Writing the average unit price of electricity different countries in kwh.

```python
for i in df1.index:
    if(df1.iloc[i , 1]=="Mainland China"):
        df1.iloc[i , 5] = 0.098
    elif(df1.iloc[i , 1]=="Russian Federation"):
        df1.iloc[i , 5] = 0.088
    elif(df1.iloc[i , 1]=="United States"):
        df1.iloc[i , 5] = 0.1042
    elif(df1.iloc[i , 1]=="Malaysia"):
        df1.iloc[i , 5] = 0.093
    elif(df1.iloc[i , 1]=="Iran, Islamic Rep."):
        df1.iloc[i , 5] = 0.004
    elif(df1.iloc[i , 1]=="Kazakhstan"):
        df1.iloc[i , 5] = 0.055
    elif(df1.iloc[i , 1]=="Canada"):
        df1.iloc[i , 5] = 0.179
    elif(df1.iloc[i , 1]=="Germany *"):
        df1.iloc[i , 5] = 0.3216
    elif(df1.iloc[i , 1]=="Ireland *"):
        df1.iloc[i , 5] = 0.213
    else:
        df1.iloc[i , 5] = 0.137 
         
df1
```

Calculating electricity used in kwh and twh monthly by the countries.

```python
df1["Electricity Used (kwh)"] = df1["Spend on electricity"]/df1["Average electicity price"]
df1["Electricity Used (Twh)"] = df1["Electricity Used (kwh)"]/(10**9)
df1
```



## Method 2: Using block reward, coin price and transaction fees

Reading the following csv files:

1. Bitcoin.csv ----- Bitcoin market price 
2. transaction-fees-usd.csv ------------- Bitcoin transaction fees 

```python
market_price = pd.read_csv("Bitcoin.csv")
transaction_fee = pd.read_csv("transaction-fees-usd.csv")

market_price.Date = pd.to_datetime(market_price.Date , format="%Y%m%")
transaction_fee.Timestamp = pd.to_datetime(transaction_fee.Timestamp , format="%Y%m%")
```

Extracting transaction fees of year 2020 and handling missing values

```python
DF = transaction_fee[transaction_fee['Timestamp'].dt.year == 2020].copy()
DF1 = pd.DataFrame()
DF1["Timestamp"] = pd.date_range(start ='01-01-2020', end ='31-12-2020', freq ='24H')
for i in DF.index:
    for j in DF1.index:
        if(DF.loc[i , "Timestamp"] == DF1.loc[j , "Timestamp"]):
            DF1.loc[j , "transaction-fees"] = DF.loc[i , "transaction-fees-usd"]
DF = DF1.fillna(method = "ffill")
DF = DF.fillna(method = "bfill")
DF            
```

Extracting price of Bitcoin of year 2020

```python
DF2 = market_price[market_price['Date'].dt.year == 2020].copy()
DF2 = DF2.reset_index()
DF2
```

Calculating Bitcoin average price on a particular day {avg. price = (High+Low/2)}

```python
for i in DF.index:
    for j in DF2.index:
        if(DF.loc[i , "Timestamp"]==DF2.loc[j , "Date"]):
            DF.loc[j , "Bitcoin_Price"] = (DF2.loc[j , "High"]+DF2.loc[j , "Low"])/2
DF
```

Adding column with block reward in BTC and USD

```python
d = pd.to_datetime("2020-05-11")
for i in DF.index:
    if(DF.loc[i , "Timestamp"]<=d):
        DF.loc[i , "Block_Reward(BTC)"] = 12.5
        DF.loc[i , "Block_Reward(USD)"] = 12.5*DF.loc[i , "Bitcoin_Price"]
    else:
        DF.loc[i , "Block_Reward(BTC)"] = 6.25
        DF.loc[i , "Block_Reward(USD)"] = 6.25*DF.loc[i , "Bitcoin_Price"]
DF
```

Calculating power consumption

```python
for i in DF.index:
    DF.loc[i , "Power_Consumption(Kwh)"] = (DF.loc[i , "Block_Reward(USD)"]+DF.loc[i , "transaction-fees"])/0.5
DF
```

## Method 3: Own Way

Reading the following csv files:

1. n-transactions-per-block.csv ----- Number of transaction per block
2. fees-usd-per-transaction.csv ----- Fees per transaction 

```python
T = pd.read_csv("n-transactions-per-block.csv")
T_Fees = pd.read_csv("cost-per-transaction.csv")

T.Timestamp = pd.to_datetime(T.Timestamp , format="%Y%m%")
T_Fees.Timestamp = pd.to_datetime(T_Fees.Timestamp , format="%Y%m%")
```

Extracting data of year 2020 and handling missing values

```python
T1 = T[T["Timestamp"].dt.year >= 2015].copy()
T1_Fees = T_Fees[T_Fees["Timestamp"].dt.year >= 2015].copy()

df01 = pd.DataFrame()
df01["Timestamp"] = pd.date_range(start ='01-01-2015', end ='31-12-2021', freq ='24H')
for i in df01.index:
    for j in T1.index:
        if(df01.loc[i , "Timestamp"] == T1.loc[j , "Timestamp"]):
            df01.loc[i , "No._of_Transactions_per_block"] = T1.loc[j , "n-transactions-per-block"]
            df01.loc[i , "Fee_per_Transaction(USD)"] = T1_Fees.loc[j , "cost-per-transaction"]
            
df01 = df01.fillna(method = "ffill")
df01 = df01.fillna(method = "bfill")
df01
```

- The block time of Bitcoin is 10 min and therefore total 144 blocks can be added and verified in a day.
- Calculating total transactions and total transaction fee per day.

```python
df01["Total_Transactions"] = df01["No._of_Transactions_per_block"]*144
df01["Total_Transaction_fees"] = df01["Total_Transactions"]*df01["Fee_per_Transaction(USD)"]
df01
```

Adding bitcoin market price and block reward

```python
DF5 = market_price[market_price["Date"].dt.year >=2015].copy()
DF5 = DF5[DF5["Date"].dt.year <=2021].copy()
DF5 = DF5.reset_index(drop=True)

for i in df01.index:
    if(df01.loc[i , "Timestamp"] == DF5.loc[i , "Date"]):
        df01.loc[i , "Market_price"] = (DF5.loc[i , "High"]+DF5.loc[i , "Low"])/2
    
d1 = pd.to_datetime("2012-11-28")
d2 = pd.to_datetime("2016-07-09")
d3 = pd.to_datetime("2020-05-11")

for i in df01.index:
    if(df01.loc[i , "Timestamp"]<d1):
        df01.loc[i , "Block_Reward(BTC)"] = 50
        df01.loc[i , "Block_Reward(USD)"] = 50*df01.loc[i , "Market_price"]
    elif(df01.loc[i , "Timestamp"]>=d1 and df01.loc[i , "Timestamp"]<d2):
        df01.loc[i , "Block_Reward(BTC)"] = 25
        df01.loc[i , "Block_Reward(USD)"] = 25*df01.loc[i , "Market_price"]
    elif(df01.loc[i , "Timestamp"]>=d2 and df01.loc[i , "Timestamp"]<d3):
        df01.loc[i , "Block_Reward(BTC)"] = 12.5
        df01.loc[i , "Block_Reward(USD)"] = 12.5*df01.loc[i , "Market_price"]
    else:
        df01.loc[i , "Block_Reward(BTC)"] = 6.25
        df01.loc[i , "Block_Reward(USD)"] = 6.25*df01.loc[i , "Market_price"]
df01
```

Calculating total revenue and power consumption
- Assuming that 60% of the revenue was spend on electricity
- Taking the average unit price 0.05 (USD/Kwh)

```python
df01["Total_revenue"]  = df01["Total_Transaction_fees"] + df01["Block_Reward(USD)"]
df01["Power_Consumption"] = df01["Total_revenue"]*(.6)/(0.05)
df01
```

## Carbon footprint

### Uisng old method

```python
Bitcoin_CF = Annual_PC.copy()
Bitcoin_CF["Footprint (gCO2/Kwh)"] = Bitcoin_CF["Power_Consumption"]*(10**9)*(475)
Bitcoin_CF
```

### New old method

**Calculating lower and upper emission values**

```python
emissions = pd.read_csv("emission.csv")
UpperBound = emissions.loc[:3]
LowerBound = emissions.loc[4:]
up = sum(UpperBound["emissions(gCO2 per kwh)"])/len(UpperBound)
lw = sum(LowerBound["emissions(gCO2 per kwh)"])/len(LowerBound)
```

**Calculating min, max and avg value of carbon emission**

```python
CarbonEmission = Annual_PC.copy()
CarbonEmission["Min Carbon Emission (Mt)"] = (CarbonEmission["Power_Consumption"]*lw)/10**3
CarbonEmission["Max Carbon Emission (Mt)"] = (CarbonEmission["Power_Consumption"]*up)/10**3
CarbonEmission["Avg Carbon Emission (Mt)"] = (CarbonEmission["Min Carbon Emission (Mt)"] + CarbonEmission["Max Carbon Emission (Mt)"])/2
CarbonEmission
```

```python
print("Carbon emission ranges from", sum(CarbonEmission["Min Carbon Emission (Mt)"]), "Mt to", sum(CarbonEmission["Max Carbon Emission (Mt)"]), "Mt.")
```
